```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../datasets/one_hot_encoded/combined-data.csv').groupby("Year").mean()


```

```python
# Plot Co2 vs Temperature as a boxplot

fig, ax = plt.subplots()

ax.boxplot([data["Co2"], data['Temperature']])
# ax.scatter(data["Co2"], data['Temperature'], color='red', label='Temperature')
ax.set_xlabel('Co2')
ax.set_ylabel('Temperature')

ax.set_title("Correlation of Co2 and Temperature")
```
